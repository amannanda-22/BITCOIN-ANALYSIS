# Bitcoin Sentiment × Hyperliquid Trader Performance

**Professional Data Science Analysis**

This notebook provides an interactive walkthrough of the full analysis.
All charts are saved to `../outputs/charts/` and the report to `../outputs/report.md`.


```python
# Add parent directory to path so we can import analysis.py
import sys
from pathlib import Path

sys.path.insert(0, str(Path('..').resolve()))

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib
matplotlib.use('Agg')   # use 'inline' for Jupyter display
import matplotlib.pyplot as plt
import seaborn as sns

print('Libraries loaded successfully.')
```

## Step 1 — Load & Clean Data


```python
from analysis import load_fear_greed, load_trader_data, merge_datasets, print_eda_summary

fg     = load_fear_greed()
td     = load_trader_data()
merged, dropped = merge_datasets(fg, td)

print_eda_summary(fg, td, merged)
```

## Step 2 — Sentiment Distribution


```python
from analysis import plot_sentiment_pie, plot_sentiment_timeline

sent_dist = plot_sentiment_pie(fg)
plot_sentiment_timeline(fg)

print('\nSentiment Counts:')
for cls, v in sent_dist.items():
    print(f'  {cls:<16}: {v["count"]:>5} days ({v["pct"]:.1f}%)')
```

## Step 3 — Trader Performance by Sentiment


```python
from analysis import (
    plot_pnl_by_sentiment, plot_winrate,
    plot_volume, plot_leverage
)

pnl_stats = plot_pnl_by_sentiment(merged)
wr        = plot_winrate(merged)
vol       = plot_volume(merged)
pos_size  = plot_leverage(merged)

print('\nPerformance Summary:')
for cls in ['Fear', 'Greed']:
    print(f'  {cls}: AvgPnL=${pnl_stats.get(cls,{}).get("mean",0):.4f}  '
          f'WinRate={wr.get(cls,0):.1f}%  '
          f'Vol=${vol.get(cls,0)/1e6:.2f}M  '
          f'AvgPos=${pos_size.get(cls,0):,.0f}')
```

## Step 4 — Hidden Patterns


```python
from analysis import (
    plot_top_symbols_fear, plot_top_symbols_greed,
    plot_buy_vs_sell, analyze_accounts,
    analyze_size_pnl_correlation, print_account_summary
)

fear_syms  = plot_top_symbols_fear(merged)
greed_syms = plot_top_symbols_greed(merged)
buy_sell   = plot_buy_vs_sell(merged)
size_corr  = analyze_size_pnl_correlation(merged)
accts      = analyze_accounts(merged)

print(f'Size-PnL Correlation: {size_corr:.4f}')
print_account_summary(accts)
```

## Step 5 — Advanced Insights


```python
from analysis import (
    plot_correlation_heatmap, plot_rolling_pnl,
    plot_pnl_distribution_extreme
)

plot_correlation_heatmap(merged)
plot_rolling_pnl(merged)
extreme_pnl = plot_pnl_distribution_extreme(merged)

print(f"Extreme Fear  avg PnL: ${extreme_pnl['extreme_fear']['mean']:.4f}")
print(f"Extreme Greed avg PnL: ${extreme_pnl['extreme_greed']['mean']:.4f}")
```

## Step 6 — Generate Report


```python
from analysis import generate_report

all_stats = {
    'sentiment_distribution': sent_dist,
    'pnl_by_sentiment':       pnl_stats,
    'win_rate':               wr,
    'volume':                 vol,
    'leverage':               pos_size,
    'buy_vs_sell':            buy_sell,
    'accounts':               accts,
    'extreme_pnl':            extreme_pnl,
    'size_pnl_corr':          size_corr,
    'total_trades':           len(merged),
    'dropped_trades':         dropped,
}
generate_report(all_stats)
print('Report written to outputs/report.md')
```
